```python
import pandas as pd

# Tabela de distribuição de frequencia da coluna CYL

df = df[['cyl']]

# Agrupando a frequencia de Cyl em Absoluta
frequencia = df.groupby('cyl').size().reset_index(name="Absoluta")

# Agrupando as frequencias relativas
freqRelativa = df.groupby('cyl').size().reset_index(name="Relativo")
freqRelativa['Relativo'] = freqRelativa['Relativo'] / 32


# Agruoando cyl em uma variavel tabela
tabela = df.groupby(['cyl']).sum()

# Juntando as tabelas frequência com Tabela
tabela = pd.merge(tabela, frequencia, on='cyl')
tabela = pd.merge(tabela, freqRelativa, on='cyl')

print(tabela)
```
