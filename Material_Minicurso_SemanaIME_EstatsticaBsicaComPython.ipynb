Comandos básicos:

```python
# Exemplo:
x = np.array(range(0,11))
y = x ** 2
```

```python
plt.plot(x, y, 'b') # 'b' é a cor azul (blue)
plt.xlabel('Eixo X', size=12)
plt.ylabel('Eixo Y', size=12)
plt.title('Título', size=20)
plt.show()
```

```python
# Multiplots na mesma tela:

# Opcional: ajustar tamanho dos gráficos:
#plt.subplots(figsize=(8,3))

## uso:  plt.subplot(nlinhas, ncols, numeroplot)
plt.subplot(1,2,1)
plt.plot(x, y, 'b--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
plt.show()
```

```python
# Cria uma figura vazia com especificação de tamanho e resolução:
plt.figure(figsize=(8,4), dpi=100)

# Plota o gráfico:
plt.plot(x, y, 'b')
plt.xlabel('Define o Label X')
plt.ylabel('Define o Label Y')
plt.title('Define o Título')
plt.show()
```

Objetos orientados

```python
# Cria uma figura vazia
fig = plt.figure()

# Adiciona eixos à figura
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Eixo principal (esquerda, inferior, largura, altura (faixa de 0 a 1))
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Eixos inseridos (esquerda, inferior, largura, altura (faixa de 0 a 1))

# Trabalhando no eixo maior
axes1.plot(x, y, 'b')
axes1.set_xlabel('Eixo X1')   # Perceba o uso de "set_" para começar métodos
axes1.set_ylabel('Eixo Y1')
axes1.set_title('Título 1')

# Trabalhando no eixo menor
axes2.plot(y, x, 'r')
axes2.set_xlabel('Eixo X2')
axes2.set_ylabel('Eixo Y2')
axes2.set_title('Título 2')
```

*   Assimetria < 0 $⟹$ assimetria negativa (à esquerda);

## **Curtose**

Neste caso:
*   Curtose = 0 $⟹$ curva mesocúrtica (normal);
*   Curtose < 0 $⟹$ curva platicúrtica (achatada);
*   Curtose > 0 $⟹$ curva leptocúrtica (afilada).

```python
dados_bussab.kurtosis()
```
